```python
# Create an API Client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
# Building Helper Functions

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content":text}
    messages.append(assistant_message)
```

```python
# Creating a list of messages and manually call client.messages.create

messages = []

add_user_message(messages, "Write a 1 sentence description of fake video")

stream = client.messages.create(
    model = model,
    max_tokens = 1000,
    messages = messages,
    stream = True
)
for event in stream:
    print(event)
```

```python
# Creating a list of messages and Streaming Response in chunks using client.messages.stream

messages = []

add_user_message(messages, "Write a 1 sentence description of a fake database")

with client.messages.stream(
    model = model,
    max_tokens = 1000,
    messages = messages
) as stream:
    for text in stream.text_stream:
        print(text, end="")
        # pass

# To get final message of stream
# stream.get_final_message()
```
