```python
# Create an API Client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
# Building Helper Functions

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content":text}
    messages.append(assistant_message)

def chat(messages, system=None, temperature=1.0, stop_sequences=[]):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
        "stop_sequences": stop_sequences
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
# Message Pre-Filling - No Longer Supported in Claude Model, though still supported on Older Claude model.

# Alternate Method - Using System Prompts or Structured Output. For Example - 

# system = "You are a TechCorp assistant. Always begin responses with 'As a TechCorp assistant...'"
# messages = [
#     {"role": "user", "content": "Describe your product plans"}
# ]

messages = []

add_user_message(messages, "Is Tea or Coffee better at Breakfast")
add_assistant_message(messages, "Tea is better because")

answer = chat(messages)

answer
```

```python
# Using Stop Sequences

messages = []

add_user_message(messages, "Count from 1 to 10")

answer = chat(messages, stop_sequences=[", 5"])

answer
```
