```python
# Create an API Client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
# Building Helper Functions

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content":text}
    messages.append(assistant_message)

# Adding Temperature Parameter: 
# The temperature parameter controls the randomness of the model's output.
# A higher temperature (e.g., 0.8) will produce more creative and diverse responses,
# while a lower temperature (e.g., 0.2) will generate more focused and deterministic answers.

def chat(messages, system=None, temperature = 1.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message.content[0].text
```

```python
# Response with Temperature set as 0

messages = []

add_user_message(messages, "Generate a one sentence movie idea")

# Low temperature - more predictable
answer = chat(messages, temperature = 0.0)

# High temperature - more predictable
# answer = chat(messages, temperature = 1.0)

answer
```
